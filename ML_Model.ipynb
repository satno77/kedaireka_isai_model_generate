```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test, z_train, z_test = train_test_split(X, y,z, test_size=0.2, random_state=42)
```

#SVR Model

```python
from sklearn.svm import SVR

model_svr_rbf = SVR(kernel="rbf")
model_svr_linear = SVR(kernel="linear")
model_svr_poly = SVR(kernel="poly")

model_svr_rbf.fit(X_train, y_train)
print("score svr kernel rbf: ", model_svr_rbf.score(X_test, y_test))

model_svr_linear.fit(X_train, y_train)
print("score svr kernel linear: ", model_svr_linear.score(X_test, y_test))

model_svr_poly.fit(X_train, y_train)
print("score svr kernel poly: ", model_svr_poly.score(X_test, y_test))

```

```python
from sklearn.metrics import mean_squared_error
import math

y_pred = model_svr_rbf.predict(X_test)
MSE = mean_squared_error(y_test, y_pred)

RMSE = math.sqrt(MSE)
print("Root Mean Square Error SVR Kernel RBF:", RMSE)

mape = np.mean(np.abs((y_test - y_pred)/y_test))*100
print("MAE(%)",round(mape,2),"%")
```

```python
y_pred = model_svr_linear.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

RMSE = math.sqrt(MSE)
print("Root Mean Square Error SVR Kernel linear:", RMSE)
mape = np.mean(np.abs((y_test - y_pred)/y_test))*100
print("MAE(%)",round(mape,2),"%")
```

```python
y_pred = model_svr_poly.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

RMSE = math.sqrt(MSE)
print("Root Mean Square Error SVR Kernel Poly:", RMSE)
mape = np.mean(np.abs((y_test - y_pred)/y_test))*100
print("MAE(%)",round(mape,2),"%")
```
