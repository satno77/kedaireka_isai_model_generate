```python
import skimage.feature as feature
#Texture dengan GLCM
for i in range(len(df_imagedir_carbonvalues)):
  image_spot = cv2.imread(df_imagedir_carbonvalues.iloc[i,0])
  gray = cv2.cvtColor(image_spot, cv2.COLOR_BGR2GRAY)
  graycom = feature.greycomatrix(gray, [1], [0, np.pi/4, np.pi/2, 3*np.pi/4], levels=256)

  # Find the GLCM properties
  contrast = feature.greycoprops(graycom, 'contrast')
  dissimilarity = feature.greycoprops(graycom, 'dissimilarity')
  homogeneity = feature.greycoprops(graycom, 'homogeneity')
  energy = feature.greycoprops(graycom, 'energy')
  correlation = feature.greycoprops(graycom, 'correlation')
  ASM = feature.greycoprops(graycom, 'ASM')

  df_imagedir_carbonvalues.at[i, 'Contrast'] = contrast.flatten()
  df_imagedir_carbonvalues.at[i, 'Dissimilarity'] = dissimilarity.flatten()
  df_imagedir_carbonvalues.at[i, 'Homogeneity'] = homogeneity.flatten()
  df_imagedir_carbonvalues.at[i, 'Energy'] = energy.flatten()
  df_imagedir_carbonvalues.at[i, 'Correlation'] = correlation.flatten()
  df_imagedir_carbonvalues.at[i, 'ASM'] = ASM.flatten()


  # print("Contrast: {}".format(contrast))
  # print("Dissimilarity: {}".format(dissimilarity))
  # print("Homogeneity: {}".format(homogeneity))
  # print("Energy: {}".format(energy))
  # print("Correlation: {}".format(correlation))
  # print("ASM: {}".format(ASM))
```

```python
# Open the image using Rasterio
for i in range(len(df_imagedir_carbonvalues)):
  with rasterio.open(df_imagedir_carbonvalues.iloc[i,0]) as src:
    red = src.read(1)
    green = src.read(2)
    nir = src.read(3)

    ndvi = ((nir.astype(float)-red)/(nir.astype(float)+red))
    ngrdi = (green.astype(float)-red)/(green.astype(float)+red)
    gndvi = (nir.astype(float)-green)/(nir.astype(float)+green)
    ci = (red-green.astype(float))/(red+green.astype(float))

    df_imagedir_carbonvalues.at[i, 'NDVI'] = ndvi.flatten()
    df_imagedir_carbonvalues.at[i, 'NGRDI'] = ngrdi.flatten()
    df_imagedir_carbonvalues.at[i, 'GNDVI'] = gndvi.flatten()
    df_imagedir_carbonvalues.at[i, 'CI'] = ci.flatten()

```
